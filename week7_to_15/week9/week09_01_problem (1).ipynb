### Q1. 현지와 수범이는 강아지와 고양이를 분류하는 binary classification 모델을 학습시켜 최종적으로 98%의 정확도를 얻었습니다. 사용한 데이터셋은 고양이 데이터가 50개, 강아지 데이터가 5000개였습니다. **이 때, 본 모델의 어떤 문제가 있는지 서술하고 해결 방안을 제시하세요.**

Your Answer :

- 문제: 강아지 데이터에 비해 고양이 데이터 수가 매우 적습니다. 데이터 불균형이 심한 데이터셋은 모델이 학습할 때 한쪽 클래스(여기서는 강아지 클래스)에만 더 많은 중요도를 둘 수 있고, 데이터가 부족한 한쪽 클래스(고양이 클래스)에 대한 패턴을 학습하기 어렵습니다. 따라서 이 예에서는 대부분의 입력을 강아지로 예측하면서 정확도가 높게 나오는 문제가 발생할 수 있습니다.

- 해결방안
  1. 고양이 클래스의 데이터를 증강하여 불균형을 완화할 수 있습니다.
  2. 모델 학습시에 강아지와 고양이 클래스에 대해 서로 다른 가중치를 부여하여 불균형을 완화할 수 있습니다.

### Q2. 표를 참고하여 Precision, Recall, Accuracy를 구체적으로 서술하세요.table1.png

Your Answer :

- Precision(정밀도): 모델이 True로 예측한 샘플 중에서 실제로 True가 맞는 비율을 나타냅니다. (모델이 Positive로 예측한 것 중에서 실제로 Positive가 맞는 비율) Precision은 True Positive/True Positive + False Positive 로 구합니다.
- Recall(재현율): 실제 True인 샘플 중에서 모델이 True로 예측한 비율입니다.
- Accuracy(정확도): 모델이 전체 샘플 중에서 알맞게 예측한 비율을 나타냅니다.

### Q3. 새 데이터 세트에서 50개의 양성 샘플과 50개의 음성 샘플이 있습니다. 이 데이터 세트를 사용하여 모델의 일반화 능력을 평가하는 방법을 제시하고, 왜 이러한 접근 방식이 중요한지 설명하세요.

Your Answer :

- 모델의 일반화 능력을 평가하기 위해서는 교차검증(corss-validation)을 사용할 수 있습니다. cross-validation에서는 데이터를 여러개의 fold로 나누어 각각 train, test 세트로 사용해 모델을 반복적으로 학습하고 평가합니다.
- 10-fold교차 검증을 사용해 데이터를 10개의 fold로 나누고, 각 반복에서는 9개의 fold가 훈련 데이터로 사용되고 1개 fold가 테스트 데이터로 사용됩니다.
- 이러한 접근방식이 중요한 이유는 먼저 overfitting을 검증할 수 있다는 것, 모델을 여러번 평가하면서 데이터에 포함된 노이즈나 이상치에 대한 영향을 평가할 수 있다는 것, 그리고 모델의 하이파라미터를 조정하거나 다른 모델들과 비교하여 성능을 평가할 수 있다는 것 등이 있습니다. 또 무엇보다 여러번의 교차검증을 통해 얻은 평균 성능은 단일 평가 결과보다 신뢰할 수 있다는 장점이 있습니다.
